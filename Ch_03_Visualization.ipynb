Dot Plot

```python
# Book Figure 3-10

import pandas as pd

!wget -O FuelConsumption.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/FuelConsumptionCo2.csv

df = pd.read_csv("FuelConsumption.csv")

# Dot Plot
values, counts = np.unique(df.CYLINDERS, return_counts=True)
# Draw dot plot with appropriate figure size, marker size and y-axis limits
fig, ax = plt.subplots(figsize=(6, 2.25))
for value, count in zip(values, counts):
    ax.plot([value]*count, list(range(count)), '', ms=2, linestyle='dotted')
for spine in ['top', 'right', 'left']:
    ax.spines[spine].set_visible(False)
ax.yaxis.set_visible(False)
ax.set_xlabel('CYLINDERS')
ax.set_ylim(-1, max(counts))
ax.set_xticks(range(min(values), max(values)+1))
ax.tick_params(axis='x', length=0, pad=8, labelsize=12)
plt.show()
```

Scatter Plot

```python
# Book Figures 3-13 to 3-17

import pandas as pd
import seaborn as sn
import matplotlib as plt
!wget -O FuelConsumption.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/FuelConsumptionCo2.csv

df = pd.read_csv("FuelConsumption.csv")

# Scatter plot
df.plot.scatter(x='ENGINESIZE', y='CO2EMISSIONS') #two dimensions/variables
df.plot.scatter(x='ENGINESIZE', y='CO2EMISSIONS',c='CYLINDERS',colormap='viridis') #three dimensions/variables with color
df.plot.scatter(x='ENGINESIZE', y='CO2EMISSIONS', s=df["CYLINDERS"]*10) #three dimensions/variables with point size
plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
sn.jointplot(x=df['ENGINESIZE'], y=df['CO2EMISSIONS'])#plot scatter with histogram
sn.pairplot(df)#plot scatter of all pairs of data

```
